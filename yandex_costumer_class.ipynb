```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
import tensorflow as tf

# Load the data
data = pd.read_csv('/content/gym_data.csv')
data=data.iloc[:,0:-1]
data=data.drop(['Row','Zipcode','Registration' ], axis=1)
X=data.drop(['Exited'],axis=1)
#data=np.mean(data,axis=1)
# Normalize
X = (X-np.mean(X,axis=0))/np.std(X)
data
#print(X)
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, data['Exited'], test_size=0.2, random_state=42)
```

```python

from sklearn.metrics import f1_score

def f1_metric(y_true, y_pred):
    y_pred = tf.round(y_pred)
    return tf.py_function(f1_score, (y_true, y_pred), tf.float32)
```

```python
y_train
```

```python

l0 = tf.keras.layers.Dense(units=4, input_shape=[8])
l1 = tf.keras.layers.Dense(units=16)
l2 = tf.keras.layers.Dense(units=16)
l3 = tf.keras.layers.Dense(64, activation='relu')
l4 = tf.keras.layers.Dense(2)#,activation='sigmoid')

model = tf.keras.Sequential([l0, l1, l2,l3,l4])
print('\n start *************************\n')
model.compile(optimizer='adam',  loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True) ,metrics=['accuracy'])#,f1_metric])#'accuracy',tf.keras.metrics.Precision(), tf.keras.metrics.Recall()])

history = model.fit(X_train, y_train, epochs=50, verbose=True)
```

in case to display the f1score, the last dense layer will have one unit and we will include the sigmoid activation & loss=tf.keras.losses.BinaryCrossentropy

```python

l0 = tf.keras.layers.Dense(units=4, input_shape=[8])
l1 = tf.keras.layers.Dense(units=16)
l2 = tf.keras.layers.Dense(units=16)
l3 = tf.keras.layers.Dense(64, activation='relu')
l4 = tf.keras.layers.Dense(1,activation='sigmoid')

model1 = tf.keras.Sequential([l0, l1, l2,l3,l4])
print('\n start *************************\n')
model1.compile(optimizer='adam',  loss=tf.keras.losses.BinaryCrossentropy(from_logits=True) ,metrics=['accuracy',f1_metric])#'accuracy',tf.keras.metrics.Precision(), tf.keras.metrics.Recall()])

history1 = model1.fit(X_train, y_train, epochs=50, verbose=True)
```
